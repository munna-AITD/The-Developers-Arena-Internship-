### Contract Type and Payment Method Together

```python
contract_payment = pd.crosstab(
    [df['Contract'], df['PaymentMethod']],
    df['Churn'],
    normalize='index'
) * 100

print(contract_payment)
```

### Contract Type
* Month-to-month customers have the highest churn risk.
* One-year contracts significantly reduce churn.
* Two-year contracts have the lowest churn risk.
### Payment Method
* Electronic check users churn most frequently.
* Automatic bank transfer and automatic credit card users churn least.
#### Recommended Actions
* Target month-to-month customers with retention campaigns.
* Offer incentives to upgrade to annual contracts.
* Encourage automatic payment enrollment.
* Monitor customers who use electronic checks, as they are often a high-risk segment.

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(
    df['Contract'],
    df['Churn']
)

chi2, p, dof, expected = chi2_contingency(table)

print("p-value:", p)
```

* P-Value <0.05. therefore Contract and Churn is significantly related

```python
table= pd.crosstab(df['PaymentMethod'],
                  df['Churn'])
chi2, p,dof, expected = chi2_contingency(table)
print("p-value", p)
```

* p-value<0.05 therefore Paymentmethod is significant related to Churn

### Which services or features (for example, online security, tech support,streaming  services, multiple lines) are most strongly associated with higher or lower churn rates?

```python
df.columns
```

```python
services = [
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies',
    'MultipleLines'
]

for service in services:
    print("\n", service)

    churn_rate = (
        df.groupby(service)['Churn']
          .apply(lambda x: (x=='Yes').mean()*100)
    )

    print(churn_rate)
```

### Create a Summary Table

```python
results = []

for service in services:

    service_rates = (
        df.groupby(service)['Churn']
          .apply(lambda x:(x=='Yes').mean()*100)
    )

    results.append([
        service,
        service_rates.max(),
        service_rates.min()
    ])

summary = pd.DataFrame(
    results,
    columns=[
        'Service',
        'Highest_Churn',
        'Lowest_Churn'
    ]
)

print(summary)
```

#C) Model Building:
a. Build a sequential model using Keras, to find out if the customerwouldchurn or not, using
‘tenure’ as the feature and ‘Churn’ as the dependent/target column:

```python
import tensorflow as tf
from tensorflow.keras import models, layers, Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
x = df[['tenure']]
y = df[['Churn']]
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
scaler= StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = Sequential([
    layers.Dense(12, activation = 'relu', input_shape = (x_train.shape[1],)),
    layers.Dense(8, activation = 'relu'),
    layers.Dense(1, activation = 'sigmoid')
])
```

```python
models.summary()
```

```python
models.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = models.fit(x_train, y_train, epochs= 150, batch_size= 32, validation_split = 0.2)
```

```python
y_pred=models.predict(x_test)
```

```python
accuracy_score(y_test, y_pred.round())
```

```python
cm = confusion_matrix(y_test, y_pred.round())
sns.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
#Visualization of accuracy and loss
plt.figure(figsize = (8,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label = 'Training Accuracy')
plt.plot(history.history['val_accuracy'], label = 'Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label = 'Training loss')
plt.plot(history.history['val_loss'], label = 'Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation loss')
plt.tight_layout()
```

```python
models.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = models.fit(x_scaled, y_train, epochs = 150, batch_size = 16, validation_split = 0.2)
```

```python
y_pred = models.predict(x_scaled_test)
```
